# Step 2: Exploratory Data Analysis (EDA)
- Explore the relationships between the target and independent variables
- Create at least five charts to show the relationships, using different chart types that are approriate to the analysis
- Ensure that all charts are clearly labeled and formatted

```python
# Chart 1 ðŸ”§
# Chart 1 - Distribution of Product purchases
plt.figure(figsize=(10, 6))
product_counts = df['product'].value_counts().sort_index()
product_labels = ['No Answer', 'No Purchase', 'Beginner', 'Intermediate', 'Advanced']
plt.bar(range(len(product_counts)), product_counts.values)
plt.xlabel('Product Type')
plt.ylabel('Count')
plt.title('Distribution of Product Purchases')
plt.xticks(range(len(product_counts)), product_labels)
for i, v in enumerate(product_counts.values):
    plt.text(i, v + 50, str(v), ha='center')
plt.tight_layout()
plt.show()
```

```python
# Chart 2 ðŸ”§
# Chart 2 - Income distribution by Product type
plt.figure(figsize=(12, 6))
df_boxplot = df[df['product'] != 0]  # Exclude no-answer customers
sns.boxplot(data=df_boxplot, x='product', y='income')
plt.xlabel('Product Type')
plt.ylabel('Income (USD)')
plt.title('Income Distribution by Product Purchase')
plt.xticks([0, 1, 2, 3], ['No Purchase', 'Beginner', 'Intermediate', 'Advanced'])
plt.tight_layout()
plt.show()
```

```python
# Chart 3 ðŸ”§
# Chart 3 - Age distribution by Product type
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Age Distribution by Product Type')

for i, product_type in enumerate([1, 2, 3, 4]):
    row, col = i // 2, i % 2
    age_data = df[df['product'] == product_type]['age']
    axes[row, col].hist(age_data, bins=20, alpha=0.7, edgecolor='black')
    product_names = ['No Purchase', 'Beginner', 'Intermediate', 'Advanced']
    axes[row, col].set_title(f'{product_names[i]} (n={len(age_data)})')
    axes[row, col].set_xlabel('Age')
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Chart 4 ðŸ”§
# Chart 4 - Job level distribution by Product
plt.figure(figsize=(12, 8))
job_product_crosstab = pd.crosstab(df['job'], df['product'], normalize='index') * 100
job_labels = ['Unemployed', 'Entry Level', 'Mid Level', 'Management/Self-employed']
product_labels = ['No Answer', 'No Purchase', 'Beginner', 'Intermediate', 'Advanced']

job_product_crosstab.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.xlabel('Job Level')
plt.ylabel('Percentage')
plt.title('Product Purchase Distribution by Job Level')
plt.xticks(range(len(job_labels)), job_labels, rotation=45)
plt.legend(title='Product', labels=product_labels, bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
# Chart 5 ðŸ”§
# Chart 5 - Checking Account vs Savings Account vs Product (Heatmap)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Checking account vs Product
chk_product = pd.crosstab(df['chk_acct'], df['product'], normalize='index') * 100
sns.heatmap(chk_product, annot=True, fmt='.1f', cmap='YlOrRd', ax=ax1)
ax1.set_title('Product Purchase Rate by Checking Account Status')
ax1.set_xlabel('Product Type')
ax1.set_ylabel('Checking Account Status')
ax1.set_xticklabels(['No Answer', 'No Purchase', 'Beginner', 'Intermediate', 'Advanced'])

# Savings account vs Product
sav_product = pd.crosstab(df['sav_acct'], df['product'], normalize='index') * 100
sns.heatmap(sav_product, annot=True, fmt='.1f', cmap='YlOrRd', ax=ax2)
ax2.set_title('Product Purchase Rate by Savings Account Status')
ax2.set_xlabel('Product Type')
ax2.set_ylabel('Savings Account Status')
ax2.set_xticklabels(['No Answer', 'No Purchase', 'Beginner', 'Intermediate', 'Advanced'])

plt.tight_layout()
plt.show()
```
