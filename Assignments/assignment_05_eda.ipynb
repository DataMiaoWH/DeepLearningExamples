## 4. Univariate Analysis

Business framing:  

Hotels care about trends like average stay length, customer mix, and pricing. A good EDA starts with understanding single variables and their distribution.

- Select at least 3 individual variables to explore
- Use plots and summary methods (e.g. info(), describe(), etc)  to describe the distribution (hint: we are only looking at the values of one variable, so think of plots you've used in the past that DON'T compare 2 variables.)
- Focus on what matters from a business standpoint (e.g., pricing, cancellations, guest types)

```python
# Your code for univariate analysis (e.g., plots, value counts) 🔧
# Variable 1: is_canceled (cancellation rate)
print("\nCancellation Distribution:")
cancellation_counts = df['is_canceled'].value_counts()
print(cancellation_counts)
print(f"Cancellation rate: {df['is_canceled'].mean():.3f}")

plt.figure(figsize=(8, 6))
df['is_canceled'].value_counts().plot(kind='bar')
plt.title('Distribution of Booking Cancellations')
plt.xlabel('Canceled (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Variable 2: adr (Average Daily Rate)
print("\nAverage Daily Rate Statistics:")
print(df['adr'].describe())

plt.figure(figsize=(10, 6))
plt.hist(df['adr'], bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Average Daily Rate (ADR)')
plt.xlabel('Average Daily Rate ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Variable 3: lead_time
print("\nLead Time Statistics:")
print(df['lead_time'].describe())

plt.figure(figsize=(10, 6))
plt.hist(df['lead_time'], bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Lead Time (Days Between Booking and Arrival)')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Variable 4: hotel type
print("\nHotel Type Distribution:")
hotel_counts = df['hotel'].value_counts()
print(hotel_counts)

plt.figure(figsize=(8, 6))
df['hotel'].value_counts().plot(kind='bar')
plt.title('Distribution of Hotel Types')
plt.xlabel('Hotel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Variable 5: market_segment
print("\nMarket Segment Distribution:")
market_counts = df['market_segment'].value_counts()
print(market_counts)

plt.figure(figsize=(12, 6))
df['market_segment'].value_counts().plot(kind='bar')
plt.title('Distribution of Market Segments')
plt.xlabel('Market Segment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 5. Bivariate Analysis

Business framing:  

Stakeholders often ask: “What drives cancellations?” or “Do longer stays mean higher revenue?” Bivariate analysis helps you uncover those kinds of relationships.

- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)
- Use scatterplots, grouped bar plots, or boxplots to explore the relationships
- Interpret what these relationships could mean for the hotel business

```python
# Your code to analyze variable relationships (e.g., scatterplots, grouped bars) 🔧
# Relationship 1: lead_time vs is_canceled
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs Cancellation Status')
plt.xlabel('Canceled (0 = No, 1 = Yes)')
plt.ylabel('Lead Time (Days)')
plt.show()

# Calculate mean lead time for canceled vs not canceled
lead_time_by_cancel = df.groupby('is_canceled')['lead_time'].mean()
print("\nMean Lead Time by Cancellation Status:")
print(lead_time_by_cancel)
```

```python
# Relationship 2: adr vs hotel type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='hotel', y='adr')
plt.title('Average Daily Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average Daily Rate ($)')
plt.show()

# Calculate mean ADR by hotel type
adr_by_hotel = df.groupby('hotel')['adr'].mean()
print("\nMean ADR by Hotel Type:")
print(adr_by_hotel)
```

```python
# Relationship 3: market_segment vs cancellation rate
cancel_by_segment = df.groupby('market_segment')['is_canceled'].mean().sort_values(ascending=False)
print("\nCancellation Rate by Market Segment:")
print(cancel_by_segment)

plt.figure(figsize=(12, 6))
cancel_by_segment.plot(kind='bar')
plt.title('Cancellation Rate by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship 4: Total stay length vs ADR
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='total_nights', y='adr', alpha=0.5)
plt.title('Total Stay Length vs Average Daily Rate')
plt.xlabel('Total Nights')
plt.ylabel('Average Daily Rate ($)')
plt.show()

# Correlation between total nights and ADR
correlation = df['total_nights'].corr(df['adr'])
print(f"\nCorrelation between total nights and ADR: {correlation:.3f}")
```
